```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
test_dataset = 'C:/chinna/DATASETS/bulray/blur/test'

# Load model
model = tf.keras.models.load_model("models\TL_MobileNetV2.h5")
model1 = tf.keras.models.load_model("models\TL_ResNet50.h5")
model3 = tf.keras.models.load_model("models\simple_cnn.h5")
model2 = tf.keras.models.load_model("models\TL_VGG16.h5")



# Create test generator
test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    test_dataset,       
    target_size=(224, 224),
    batch_size=32,
    class_mode="sparse",  
    shuffle=False
)

# Evaluate model (steps required)
print("ALL Models Are Train For The 5 epoches\n")

print("For MobileNet Transfer Learning Model")
test_loss, test_acc = model.evaluate(test_generator, steps=len(test_generator))
print("Test Loss:", test_loss)
print("Test Accuracy:", test_acc * 100)

print("\nFor Simple CNN Model")
test_loss, test_acc = model3.evaluate(test_generator, steps=len(test_generator))
print("Test Loss:", test_loss)
print("Test Accuracy:", test_acc * 100)


```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# Helper function for evaluation
# -------------------------------
def evaluate_model(model, model_name, test_generator):
    print(f"\n{model_name}")
    
    # 1. Get predictions
    y_pred_probs = model.predict(test_generator)
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = test_generator.classes

    # 2. Accuracy
    acc = accuracy_score(y_true, y_pred)
    print("Accuracy:", acc)

    # 3. Precision, Recall, F1 (per class + average)
    print("\nClassification Report:\n")
    print(classification_report(y_true, y_pred, target_names=test_generator.class_indices.keys()))

    # 4. Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=test_generator.class_indices.keys(),
                yticklabels=test_generator.class_indices.keys())
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()


# -------------------------------
# Evaluate all models
# -------------------------------
evaluate_model(model, "MobileNetV2 Transfer Learning Model", test_generator)
evaluate_model(model3, "Simple CNN Model", test_generator)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.utils import load_img, img_to_array

# Load your trained models
model = tf.keras.models.load_model("models\TL_MobileNetV2.h5")
model1 = tf.keras.models.load_model("models\TL_ResNet50.h5")
model3 = tf.keras.models.load_model("models\simple_cnn.h5")
model2 = tf.keras.models.load_model("models\TL_VGG16.h5")



# Load stored class names
%store -r class_names


def predict_with_models(image_path, models, model_names):
    # Load and resize image
    img = load_img(image_path, target_size=(224, 224))
    plt.imshow(img)
    plt.title(f"Image: {image_path.split('/')[-1]}")
    plt.axis('off')
    plt.show()

    # Prepare image
    img_array = img_to_array(img) / 255.0  
    img_array = np.expand_dims(img_array, axis=0)

    # Predict with each model
    for model, name in zip(models, model_names):
        predictions = model.predict(img_array, verbose=0)
        predicted_index = np.argmax(predictions[0])
        predicted_class = class_names[predicted_index]
        confidence = 100 * np.max(predictions[0])
        print(f"{name} → Predicted: {predicted_class} ({confidence:.2f}%)")
    print("-" * 60)


# Example usage with 3 models
models = [model, model3]
model_names = ["MobileNet", "Simple CNN"]

image_list = [
    "C:/chinna/DATASETS/bulray/blur/test/bird/1_Radial_Blur.jpg"
]

for img_path in image_list:
    predict_with_models(img_path, models, model_names)
```

```python
import cv2
import numpy as np
import tensorflow as tf
import time

# -------------------------------
# Load model
# -------------------------------
model = tf.keras.models.load_model("models/TL_MobileNetV2.h5")
class_names = ['bird', 'bus', 'car', 'cat', 'dog', 'truck']

# -------------------------------
# Prediction function
# -------------------------------
def predict_frame(frame):
    img = cv2.resize(frame, (224, 224)) / 255.0
    img_array = np.expand_dims(img, axis=0)
    preds = model.predict(img_array, verbose=0)
    predicted_index = np.argmax(preds[0])
    confidence = np.max(preds[0])
    return class_names[predicted_index], confidence

# -------------------------------
# Video classification with FPS
# -------------------------------
def classify_video(video_path):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    frame_count = 0
    start_time = time.time()

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        predicted_class, confidence = predict_frame(frame)

        # FPS calculation (live)
        frame_count += 1
        elapsed = time.time() - start_time
        fps = frame_count / elapsed if elapsed > 0 else 0

        # Overlay prediction + FPS
        cv2.putText(frame, f"{predicted_class} ({confidence*100:.2f}%)",
                    (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.putText(frame, f"FPS: {fps:.2f}", 
                    (20, 80), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)

        cv2.imshow("Video Classification", frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
    print(f"Processed {frame_count} frames → Avg FPS: {fps:.2f}")

# -------------------------------
# Run on a sample video
# -------------------------------
classify_video("test153.mp4")
```
